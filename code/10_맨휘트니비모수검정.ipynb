```python
import pandas as pd
from scipy.stats import mannwhitneyu
k=pd.read_csv("data/중극장_요인분석_유형분류_회복력조인.csv")
```

# 군집 0

```python
df0=k[k['cluster']==0]
df0
```

```python
k0=df0[df0['충격회복력']==0] # 침체기가 없는 극장
k1=df0[df0['충격회복력']==-100] # 회복을 하지 못한 극장
kbig=df0[(df0['충격회복력']!=-100)&(df0['충격회복력']!=0)&(df0['resilience']>0)] #회복탄력성 지수가 양수인 극장
ksmall=df0[(df0['충격회복력']!=-100)&(df0['충격회복력']!=0)&(df0['resilience']<0)] #회복탄력성 지수가 음수인 극장
```

```python
for i in range(1,23):
    print(ksmall.columns[i],mannwhitneyu(k1.iloc[:,i], kbig.iloc[:,i])) #회복을 하지 못한 극장과 회복탄력성 지수가 양수인 극장의 평균 비교 검정(비모수 검정)
```

```python
print(k1['예매/취소금액'].mean())
print(kbig['예매/취소금액'].mean())
```

```python
print(k1['장당금액'].mean())
print(kbig['장당금액'].mean())
```
